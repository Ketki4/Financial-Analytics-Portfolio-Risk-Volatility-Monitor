# MERGING API

```python
# Load old dataset
old_data = pd.read_csv("stock.csv")

# Fetch new data
new_data = yf.download("TCS.NS", start="2024-01-01", end="2025-01-01")

# Reset index (important)
new_data.reset_index(inplace=True)

# Combine datasets
updated_data = pd.concat([old_data, new_data])

# Remove duplicates
updated_data.drop_duplicates(subset="Date", inplace=True)

# Save updated dataset
updated_data.to_csv("updated_stock.csv", index=False)
```
